```python
def category(year):
    if year>1900 and year<=1925:
        return 'Quarter1'
    if year>1925 and year<=1950:
        return 'Quarter2'
    if year>1950 and year<=1975:
        return 'Quarter3'
    if year>1975 and year<=2000:
        return 'Quarter4'
    else:
        return 'Phase5'
```
